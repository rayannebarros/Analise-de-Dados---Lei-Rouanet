<h1> TRABALHO ESTATISTICA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h1> Questão 1

```python
df = pd.read_csv('train.csv')
df.columns = ['Area Cultural', '2017', '2018']
```

```python
df.set_index('Area Cultural')
```

```python
for i in range(len(df['2017'])):
    df['2017'].loc[i] = df['2017'].loc[i].replace(",", "")
```

```python
df['2017'] = df['2017'].astype(np.float).astype(np.int)
```

```python
df = df.sort_values(by='2017', ascending=False)
```

```python
n_df = df[0:3].drop('2018',axis=1)
n_df
```

```python
n_df[['Area Cultural', '2017']].plot.pie(y='2017', figsize=(5,5), labels=n_df['Area Cultural'])
```

```python
nn_df = n_df[['Area Cultural', '2017']]
nn_df = nn_df.set_index("""Area Cultural""")
```

```python
nn_df.plot.barh()
```

<h1> Questão 2

```python
df3 = pd.read_csv('train2.csv')
df3.columns = ['Segmento Cultural', '2017', '2018']
df3.set_index('Segmento Cultural')
```

```python
df3['2017'] = df3['2017'].astype(np.str)
for i in range(len(df3)):
    df3['2017'].loc[i] = df3['2017'].loc[i].replace(",", "")
```

```python
for i in range(len(df3)):
    df3['2017'].loc[i] = df3['2017'].loc[i].replace("nan", "0")
```

```python
df3['2017'] = df3['2017'].astype(np.float).astype(np.int)
df3
```

```python
df4 = df3.drop("2018",axis=1)
df4
```

```python
df4 = df4.sort_values(by='2017', ascending=False)
df4 = df4[:10]
df4
```

```python
df4[['Segmento Cultural', '2017']].plot.pie(y='2017', figsize=(8,8), labels=df4['Segmento Cultural'])
circulo=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circulo)
plt.legend(loc='center')
```

```python
cores=['red', 'purple', 'black','blue','yellow','brown','orange','magenta','grey', 'green']
nn_df4 = df4[['Segmento Cultural', '2017']]
nn_df4 = nn_df4.set_index("""Segmento Cultural""")
nn_df4.plot.barh(color=cores)
```
